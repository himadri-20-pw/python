```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=200, noise = 0.1)
```

```python
dbscan = DBSCAN(eps = 0.2) 
```

```python
dbscan.fit(X)
```

```python
clusters = dbscan.fit_predict(X)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[clusters != -1, 0], X[clusters != -1, 1], c=clusters[clusters != -1], cmap='viridis',marker='o', s=50, label='Clusters')
plt.scatter(X[clusters == -1, 0], X[clusters == -1, 1], c='red', marker='x', s=100, label='Outliers')

plt.title('DBSCAN Clustering of Moon-shaped Data with Outliers Highlighted')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
data = datasets.load_wine()
X = data.data
y = data.target
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
X_pca = pca.fit_transform(X)
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2)
```

```python
cluster.fit(X_pca)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster.labels_, cmap='viridis', s=50, alpha=0.8)
plt.title('Agglomerative Clustering on Wine Dataset (PCA-reduced to 2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
file_path= "C:\\Users\\KIIT\\Desktop\\clusttering\\E-commerce.csv"
data = pd.read_csv(file_path)
data
```

```python
X = data.drop('Purchase History', axis=1)
y = data['Purchase History']
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
num_cols = X.select_dtypes(include=['int64','float64']).columns
cat_cols = X.select_dtypes(include=['object']).columns
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ])
```

```python
X_scaled = preprocessor.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=1)
y = kmeans.fit_predict(X_scaled)
```

```python
from sklearn.metrics import silhouette_score
inertia = []
silhouette_scores = []
K = range(2, 10)
```

```python
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    sil_scores.append(silhouette_score(X_scaled, kmeans.labels_))
```

```python
plt.plot(K, inertia, 'bo-')
plt.xlabel("Number of clusters")
plt.ylabel("Inertia (Within-cluster sum of squares)")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
plt.plot(K, sil_scores, 'ro-')
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for Different k")
plt.show()
```
