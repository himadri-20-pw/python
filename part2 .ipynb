### Q1) What is hypothesis testing in statistics?

### Q2) What is the null hypothesis, and how does it differ from the alternative hypothesis?

#### ANS- The null hypothesis is the hypothesis to be tested for possible rejection under the assumption that it is true. In simple words it  is initial or default assumption. The concept of the null is similar to innocent until proven guilty We assume innocence until we have enough evidence to prove that a suspect is guilty.H0 denotes it.Example-person is not gulity.
#### The null hypothesis (H0) states that there is no significant relationship or difference between variables, while the alternative hypothesis (H1 or Ha) proposes that there is a relationship or difference. Null hypothesis represents the status quo or a baseline assumption to test against where as alternative hypothesis what you want to prove or support with evidence.

### Q3) What is the significance level in hypothesis testing, and why is it important?

#### ANS- In hypothesis testing, the significance level, denoted by α (alpha), is the probability of rejecting the null hypothesis when it is actually true.It's a predetermined threshold used to determine if the evidence against the null hypothesis is strong enough to be considered statistically significant.Hypothesis testing is important as :-
#### Controls Error Rate:
The significance level sets a boundary for how much risk you're willing to take in wrongly rejecting the null hypothesis. Lower α = more cautious testing.
#### Determines the Critical Value:
It helps define the critical region: the set of values for which the null hypothesis is rejected.For example, in a z-test, α determines the z-score threshold (e.g., ±1.96 for α = 0.05 in a two-tailed test).
#### Affects Statistical Power:
There’s a trade-off: lowering α reduces Type I error but increases the risk of a Type II error (failing to reject a false null hypothesis).
#### Used in p-value Comparison:
A result is considered statistically significant if the p-value < α.This helps you decide whether your test results provide enough evidence to reject the null hypothesis.

### Q5) How do you interpret the P-value in hypothesis testing?

#### ANS- In hypothesis testing, the p-value represents the probability of obtaining a result as extreme as, or more extreme than, the one observed in a sample, assuming the null hypothesis is true.It is also called the probability of observing the data (or more extreme data) if the null hypothesis were true. Essentially, it quantifies the strength of evidence against the null hypothesis.

### Q8) What is the Z-test, and when is it used in hypothesis testing?

#### ANS- Z-test is a statistical hypothesis test used to determine if there's a significant difference between a sample mean and a population mean, or between the means of two populations, when the population standard deviation is known and the sample size is large (n > 30). It relies on the normal distribution to approximate the distribution of the test statistic. Z-tests are generally used when the sample size is 30 or more (n ≥ 30), the sample mean can be reasonably approximated by a normal distribution, making the Z-test suitable.Z-tests can be used to compare a sample mean to a hypothesized population mean or to compare the means of two independent populations.  

### Q9) How do you calculate the Z-score, and what does it represent in hypothesis testing? 

#### ANS- To calculate the Z-score:-
#### 1. Identify the population mean (μ) and standard deviation (σ). 
#### 2. Find the difference between the data point (x) and the population mean (μ). 
#### 3. Divide the difference by the population standard deviation (σ).
#### formula:- z = (x - μ) / σ
#### In hypothesis testing, it represents how many standard deviations a sample statistic (like a sample mean) is away from the hypothesized population mean. This allows you to determine if the observed sample is likely to have come from the population with the hypothesized mean. If the Large Z-score , it indicates the sample statistic is far from the hypothesized population mean, suggesting a strong difference between the sample and population.If the Small Z-score, it indicates the sample statistic is close to the hypothesized population mean, suggesting little difference between the sample and population. If the Critical Values, Z-scores are compared to critical values (determined by the chosen significance level) to decide whether to reject or fail to reject the null hypothesis.

### Q3) Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

#### ANS- 

```python
def bernoulli_pdf(x, p):

    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0.0  # Any other value has probability 0 in a Bernoulli distribution
```

```python
print(bernoulli_pdf(1, 0.6))  
print(bernoulli_pdf(0, 0.6))  
print(bernoulli_pdf(2, 0.6))
```

### Q9) Write a Python function to calculate Z-scores from a dataset and plot them.

#### ANS- 

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    if std == 0:
        raise ValueError("Standard deviation is zero. Z-scores are undefined.")
    z_scores = (data - mean) / std
    plt.hist(z_scores, bins=50, density=True, alpha=0.7, color='green', edgecolor='black')
    plt.title('Histogram of Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()
    return z_scores
```

### Q14) Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

#### ANS- 

```python
import numpy as np
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
    z_score = (data_point - mean) / std_dev
    return z_score

def compare_to_standard_normal(z_score):
   
    probability = norm.cdf(z_score)
    if z_score > 0:
        description = f"The data point is {z_score:.2f} standard deviations above the mean. The probability of a value less than this is {probability:.4f}."
    elif z_score < 0:
        description = f"The data point is {abs(z_score):.2f} standard deviations below the mean. The probability of a value less than this is {probability:.4f}."
    else:
        description = "The data point is equal to the mean. The probability of a value less than this is 0.5."
    return probability, description
```

```python
data_point = 10
mean = 5
std_dev = 2

z_score = calculate_z_score(data_point, mean, std_dev)
probability, description = compare_to_standard_normal(z_score)

print(f"Z-score: {z_score:.2f}")
print(description)
```

### Q15) Implement hypothesis testing using Z-statistics for a sample dataset.

#### ANS- 

```python
import numpy as np
from scipy.stats import norm

sample_data = [105, 110, 98, 100, 102, 97, 103, 108, 99, 101]

mu_0 = 100            
sigma = 5
alpha = 0.05      
n = len(sample_data)
sample_mean = np.mean(sample_data)

z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

z_critical = norm.ppf(1 - alpha / 2)

p_value = 2 * (1 - norm.cdf(abs(z)))

print("Sample Mean:", sample_mean)
print("Z-statistic:", z)
print("Z-critical value (±):", z_critical)
print("P-value:", p_value)

if abs(z) > z_critical:
    print("Reject the null hypothesis (H₀).")
else:
    print("Fail to reject the null hypothesis (H₀).")
```

### Q16) Create a confidence interval for a dataset using Python and interpret the result.

#### ANS-

```python
import numpy as np
import scipy.stats as stats

data = [85, 87, 90, 88, 86, 91, 89, 84, 90, 92]

mean = np.mean(data)
sem = stats.sem(data)  
confidence = 0.95
h = sem * stats.t.ppf((1 + confidence) / 2.0, len(data) - 1)
lower_bound = mean - h
upper_bound = mean + h

print(f"Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

### Q24) Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

### Q1) Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

#### ANS- 

```python
import math
from scipy.stats import norm

def z_test(sample_mean, sample_std, sample_size, population_mean, alpha):
   
    z_statistic = (sample_mean - population_mean) / (sample_std / math.sqrt(sample_size))

    if z_statistic > 0:
        p_value = 2 * (1 - norm.cdf(z_statistic))
    else:
        p_value = 2 * norm.cdf(z_statistic)

    reject_null = p_value < alpha


    results_string = f"Z-statistic: {z_statistic:.3f}\nP-value: {p_value:.3f}\n"

    if reject_null:
        results_string += "Reject the null hypothesis. The sample mean is significantly different from the population mean.\n"
    else:
        results_string += "Fail to reject the null hypothesis. There is no significant difference between the sample mean and the population mean.\n"

    return z_statistic, p_value, reject_null, results_string
```

```python
sample_mean = 52  
sample_std = 8    
sample_size = 50  
population_mean = 50 
alpha = 0.05

z_stat, p_val, reject, results = z_test(sample_mean, sample_std, sample_size, population_mean, alpha)
print(results)
```

### Q5) Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

#### ANS-   

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def simple_error_plot(mu0=0, mu1=1, sigma=1, alpha=0.05):

    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)

    y0 = norm.pdf(x, mu0, sigma)
    y1 = norm.pdf(x, mu1, sigma)

    z_crit = norm.ppf(1 - alpha)
    crit_value = mu0 + z_crit * sigma
    plt.figure(figsize=(8, 5))
    plt.plot(x, y0, label='H₀: μ = μ₀', color='blue')
    plt.plot(x, y1, label='H₁: μ = μ₁', color='green')

    # Type I Error (α): Reject H₀ when true
    plt.fill_between(x, 0, y0, where=(x > crit_value), color='red', alpha=0.4, label='Type I Error (α)')

    # Type II Error (β): Fail to reject H₀ when H₁ is true
    plt.fill_between(x, 0, y1, where=(x <= crit_value), color='orange', alpha=0.4, label='Type II Error (β)')

    plt.axvline(crit_value, color='black', linestyle='--', label='Critical Value')
    plt.title('Type I and Type II Errors')
    plt.xlabel('Test Statistic')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    beta = norm.cdf(crit_value, mu1, sigma)
    power = 1 - beta
    print(f'Type II Error (β): {beta:.4f}')
    print(f'Test Power: {power:.4f}')
```

```python
simple_error_plot(mu0=0, mu1=1, sigma=1, alpha=0.05)
```

### Q9) Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

#### ANS- 

```python
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence_level=0.95):
    sample_mean = np.mean(data)
    sample_std_dev = np.std(data, ddof=1) 
    degrees_of_freedom = len(data) - 1
    t_critical = stats.t.ppf(1 - (1 - confidence_level) / 2, degrees_of_freedom)
    margin_of_error = t_critical * (sample_std_dev / np.sqrt(len(data)))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound
```

```python
sample_data = [10, 12, 15, 18, 20, 22, 25, 28]
confidence_level = 0.95
lower_bound, upper_bound = calculate_confidence_interval(sample_data, confidence_level)
print(f"The {confidence_level * 100}% confidence interval for the sample mean is: ({lower_bound:.2f}, {upper_bound:.2f})")
```

The significance of a confidence interval lies in its ability to quantify the uncertainty associated with an estimate and provide a range of plausible values for the population mean.

### Q10) Write a Python program to calculate the margin of error for a given confidence level using sample data.

#### ANS- 

```python
import numpy as np
import scipy.stats as st
import math

def calculate_margin_of_error(data, confidence_level):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  
    sample_size = len(data)
    
    alpha = 1 - confidence_level
    degrees_of_freedom = sample_size - 1

    critical_value = st.t.ppf(1 - alpha/2, degrees_of_freedom)
    standard_error = sample_std / math.sqrt(sample_size)
    margin_of_error = critical_value * standard_error

    return margin_of_error
```

```python
sample_data = [72, 78, 68, 80, 75, 71, 77, 73, 70, 74]
confidence_level = 0.95
margin = calculate_margin_of_error(sample_data, confidence_level)
print(f"Margin of error: {margin:.3f}")
```

### Q11) Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

#### ANS- Bayesian inference in Python involves using Bayes' theorem to update probabilities based on new evidence. This process starts with a prior probability representing initial belief, a likelihood representing the probability of observing evidence given the hypothesis, and a posterior probability representing the updated belief after considering the evidence.

```python
hypotheses = {
    'fair': 0.5,
    'biased': 0.9
}
priors = {
    'fair': 0.5,
    'biased': 0.5
}

# Observed data
heads = 9
tails = 1
def likelihood(hypothesis_p, heads, tails):
    return (hypothesis_p * heads) * ((1 - hypothesis_p) * tails)

unnormalized_posteriors = {}
for hypothesis, p in hypotheses.items():
    prior = priors[hypothesis]
    like = likelihood(p, heads, tails)
    unnormalized_posteriors[hypothesis] = prior * like

evidence = sum(unnormalized_posteriors.values())
posteriors = {h: up / evidence for h, up in unnormalized_posteriors.items()}

print("Posterior probabilities:")
for h, p in posteriors.items():
    print(f"{h}: {p:.4f}")
```

#### process:-
##### Define Hypotheses: We assume two possible truths about the coin.
##### Assign Priors: We start with equal belief in both.
##### Compute Likelihood: For each hypothesis, calculate the likelihood of observing 9 heads and 1 tail.
##### Apply Bayes' Theorem:
##### Multiply prior and likelihood to get unnormalized posteriors.
##### Normalize these to get actual posterior probabilities.
##### Interpret: The result tells you which hypothesis is more likely given the data.

### Q12) Perform a Chi-square test for independence between two categorical variables in Python.

#### ANS-

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
 
data = {'Category1': ['A', 'A', 'B', 'B', 'A', 'B', 'A', 'B'],
           'Category2': ['X', 'Y', 'X', 'Y', 'X', 'X', 'Y', 'Y']}
df = pd.DataFrame(data)

  
contingency_table = pd.crosstab(df['Category1'], df['Category2'])
print("Contingency Table:\n", contingency_table)

chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:\n", expected)

alpha = 0.05
print("P-value:", p)
if p <= alpha:
       print("Reject H0: The variables are dependent.")
else:
       print("Fail to reject H0: The variables are independent.")
```

### Q13) Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

#### ANS-

```python
import numpy as np
def calculate_expected_frequencies(observed_data):

    observed_data = np.array(observed_data)
    row_totals = observed_data.sum(axis=1)
    col_totals = observed_data.sum(axis=0)
    total = observed_data.sum()
    expected_frequencies = np.outer(row_totals, col_totals) / total
    return expected_frequencies
```

```python
observed_data = [
        [20, 30, 50],
        [10, 40, 20],
        [30, 20, 10]
    ]
expected_frequencies = calculate_expected_frequencies(observed_data)
print("Observed Frequencies:")
print(np.array(observed_data))
print("\nExpected Frequencies:")
print(expected_frequencies)
```

### Q14) Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

#### ANS-

```python
import numpy as np
from scipy import stats

observed_frequencies = np.array([10, 12, 8, 15, 14])

total_count = sum(observed_frequencies)
expected_frequencies = np.array([total_count / len(observed_frequencies)] * len(observed_frequencies))

statistic, p_value = stats.chisquare(observed_frequencies, expected_frequencies)

print("Chi-squared statistic:", statistic)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
  print("Reject the null hypothesis (observed data significantly different from expected)")
else:
  print("Fail to reject the null hypothesis (observed data not significantly different from expected)")
```
